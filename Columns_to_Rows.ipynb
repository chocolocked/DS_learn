```python
df = pd.read_csv('earthquake.csv')
df.head()
```

```python
#1). using pd.melt function 
new_df=pd.melt(df, id_vars=['magnitude'],
               value_vars=['year_2000','year_2001','year_2002','year_2003','year_2004','year_2005', 'year_2006','year_2007','year_2008','year_2009','year_2010','year_2011','year_2012'],
               var_name='year', value_name='number_of_earthq')
```

```python
#clean the Year column
new_df['year']=new_df['year'].str.replace('year_','')
```

```python
#2)pandas sql 
pysqldf = lambda q: pdsql.sqldf(q, globals())
```

```python
# run any valid sql query
pysqldf('select * from df limit(5)')
```

```python
# hardcoded values
pysqldf("""
SELECT years.*
    FROM 
    (   SELECT 2000 AS year
        UNION SELECT 2001 UNION SELECT 2002 UNION SELECT 2003
        UNION SELECT 2004 UNION SELECT 2005 UNION SELECT 2006
        UNION SELECT 2007 UNION SELECT 2008 UNION SELECT 2009
        UNION SELECT 2010 UNION SELECT 2011 UNION SELECT 2012
    )   years
""")
```

```python
#concatenate cross product of years and all df rows
pysqldf("""
SELECT
    test.year as year,
    df.*
FROM df
CROSS JOIN
(SELECT years.*
        FROM 
    (   SELECT 2000 AS year
        UNION SELECT 2001 UNION SELECT 2002 UNION SELECT 2003
        UNION SELECT 2004 UNION SELECT 2005 UNION SELECT 2006
        UNION SELECT 2007 UNION SELECT 2008 UNION SELECT 2009
        UNION SELECT 2010 UNION SELECT 2011 UNION SELECT 2012
    ) years
 ) test 
""")
```

```python
#lastly,do the case when to modify the table
pysqldf("""
SELECT a.year,
       a.magnitude,
       CASE WHEN a.year = 2000 THEN a.year_2000 
            WHEN a.year = 2001 THEN a.year_2001
            WHEN a.year = 2002 THEN a.year_2002
            WHEN a.year = 2003 THEN a.year_2003
            WHEN a.year = 2004 THEN a.year_2004
            WHEN a.year = 2005 THEN a.year_2005
            WHEN a.year = 2006 THEN a.year_2006
            WHEN a.year = 2007 THEN a.year_2007
            WHEN a.year = 2008 THEN a.year_2008
            WHEN a.year = 2009 THEN a.year_2009
            WHEN a.year = 2010 THEN a.year_2010
            WHEN a.year = 2011 THEN a.year_2011
            WHEN a.year = 2012 THEN a.year_2012
            ELSE NULL END AS number_of_earthq
FROM
(SELECT
    test.year as year,
    df.*
FROM df
CROSS JOIN
(SELECT years.*
        FROM 
    (   SELECT 2000 AS year
        UNION SELECT 2001 UNION SELECT 2002 UNION SELECT 2003
        UNION SELECT 2004 UNION SELECT 2005 UNION SELECT 2006
        UNION SELECT 2007 UNION SELECT 2008 UNION SELECT 2009
        UNION SELECT 2010 UNION SELECT 2011 UNION SELECT 2012
    ) years
 ) test
 ) a 
""")
```
